```r
library(ggplot2)
library(grid)
library(gridExtra)

jp_data <- read.csv2("jp_b_d.csv")
old_jp_data <- jp_data[1:10,]
new_jp_data <- jp_data[11:nrow(jp_data),]
#new_jp_death_ts <- ts(new_jp_data$Death, start=c(1966), end=c(2014), frequency=1)
#new_jp_birth_ts <- ts(new_jp_data$Birth, start=c(1966), end=c(2014), frequency=1)

#old_jp_death_ts <- ts(old_jp_data$Death, start=c(1776), end=c(1956), frequency=.05)
#old_jp_birth_ts <- ts(old_jp_data$Birth, start=c(1776), end=c(1956), frequency=.05)

#ts.plot(old_jp_death_ts, old_jp_birth_ts, new_jp_death_ts, new_jp_birth_ts, gpars= list(col = c("blue", "red", "blue", "red")))


g1 <- ggplot(data = jp_data, aes(x=Year)) +
    geom_vline(xintercept = 1856, linetype = "longdash") +
    geom_vline(xintercept = 1967, linetype = "longdash") +
    geom_vline(xintercept = 1990, linetype = "longdash") +
    geom_vline(xintercept = 2004, linetype = "longdash") +
    annotation_custom(grob = textGrob("Steg 2", 
                                      gp = gpar(fontsize = 9, col = 'black'),),  
        xmin = 1856, xmax = 1856, ymin = 2.3, ymax = 2.35) +
    annotation_custom(grob = textGrob("Steg 3", 
                                      gp = gpar(fontsize = 9, col = 'black'),),  
        xmin = 1967, xmax = 1967, ymin = 2.3, ymax = 2.35) +
    annotation_custom(grob = textGrob("Steg 4", 
                                      gp = gpar(fontsize = 9, col = 'black'),),  
        xmin = 1990, xmax = 1990, ymin = 2.3, ymax = 2.35) +
    annotation_custom(grob = textGrob("Steg 5", 
                                      gp = gpar(fontsize = 9, col = 'black'),),  
        xmin = 2006, xmax = 2006, ymin = 2.3, ymax = 2.35) +
    geom_line(aes(y = Death, color = "red")) +
    geom_line(aes(y = Birth, color = "blue")) +
    scale_color_manual("",labels = c("Nativitet", "Mortalitet"), values = c("#56B4E9", "#D55E00")) +
    ylab("Nativitet / Mortalitet Per 1000") +
    xlab("") +
    theme_bw() +
    theme(legend.position="bottom", plot.margin = unit(c(1,1,3,1),"lines"))


g = ggplotGrob(g1)
g$layout$clip[g$layout$name == "panel"] = "off"
grid.draw(g)
ggsave("full.pdf", g, width = 10, height = 6)


#ggplot(data = old_jp_data, aes(x=Year)) +
#    geom_line(aes(y = Death, color = "red")) +
#    geom_line(aes(y = Birth, color = "blue")) +
#    scale_color_manual("",labels = c("Death", "Birth"), values = c("red", "blue")) +
#    theme_bw()

```
